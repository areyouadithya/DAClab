### Question 11

```python
for i in cars['company'].unique():
    print(f"For company {i}:")
    print("Mean horsepower:",cars[cars['company']==i]['horsepower'].mean())
    print("Median horsepower:",cars[cars['company']==i]['horsepower'].median())
    print("Mode horsepower:",cars[cars['company']==i]['horsepower'].mode())
    print("\n")
```

### Question 12

```python
for i in cars['company'].unique():
    print("For company",i,":")
    avgmile = cars[cars['company']==i]['average-mileage']
    price = cars[cars['company']==i]['price']
    print("Covariance between average-mileage and price:",avgmile.cov(price))
    print("Correlation between average-mileage and price:",avgmile.corr(price))
    print("\n")
```

### PCA on horsepower and average-mileage

```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA

X = pd.read_csv("cars.csv",usecols=['horsepower','average-mileage'])
XX = pd.read_csv("cars.csv",usecols=['company'])
y=XX['company']
target_names = cars['company'].unique()

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

plt.figure()
colors = ["navy", "turquoise", "darkorange","red","yellow","black"]
lw = 2

for color, i, target_name in zip(colors, [0,1,2,3,4,5], target_names):
    plt.scatter(
        X_r[y == i, 0], X_r[y == i, 1], color=color, alpha=0.8, lw=lw, label=target_name
    )
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("PCA")
```
