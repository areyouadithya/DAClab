```python
import pandas as pd
from sklearn.decomposition import PCA
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('iris.csv')
df['Total'] = df['sepal.length'] + df['petal.width'] + df['petal.length'] + df['sepal.width']
X = df[['Total']]
y = df['variety']

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
pca = PCA(n_components=1)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.fit_transform(X_test)



label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)

plt.scatter(X_train_pca, X_train_pca, c=y_train_encoded)
plt.xlabel("PCA")
plt.ylabel("Arbitrary")
plt.title("PCA of Boston dataset")
plt.legend(np.unique(y_train))
plt.colorbar(label=np.unique(y_train))
plt.show()
```
