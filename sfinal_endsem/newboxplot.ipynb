```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = pd.read_csv("penguins.csv")
sns.boxplot(data)
plt.show()
new_data = data.drop(["sex", "island","species"], axis=1)
q1 = new_data.quantile(0.25)
q3 = new_data.quantile(0.75)
iqr = q3 - q1
lowerbound = q1 - 1.5*iqr
upperbound = q3 + 1.5*iqr
clean_data = new_data[~((lowerbound >= new_data) | (upperbound <= new_data) ).any(axis=1)]

sns.boxplot(clean_data)
plt.show()

```
